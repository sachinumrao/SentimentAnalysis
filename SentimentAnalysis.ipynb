```python
import numpy as np
import pandas as pd
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
import contractions
import string
```

```python
def tokenizer(text, stopwords):
    
    # Fix contractions
    fixed_text = ''+contractions.fix(text)
    
    # Remove punctuations
    imputed_text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Generate word list by splitting text
    raw_word_list = imputed_text.split(' ')
    
    # Remove numbers from list
    word_list = [x for x in raw_word_list if x.isalpha()]
    
    # Remove stopwords
    f_word_list_1 = [x for x in word_list if x not in stopwords]
    
    # Lemmatize the word list
    lemmatizer = WordNetLemmatizer()
    f_word_list_3 = [lemmatizer.lemmatize(x.lower()) for x in f_word_list_2]
    f_word_list_3 = [x.lower() for x in f_word_list_1]
    
    return f_word_list_3
```

```python
def POS_remover(word_list, filter_list):
    tagged_word_list = nltk.pos_tag(word_list)
    f_word_list = []
    for i in range(len(word_list)):
        if tagged_word_list[i][1] in filter_list:
            continue
        else:
            f_word_list.append(word_list[i])
    
    return f_word_list
```

```python
stopwords_set = set(stopwords.words('english'))
```

# Read Dataset

```python
df = pd.read_csv('dataset.csv', encoding = "ISO-8859-1")
```

# Generate Word_list

```python
raw_data = []
```

```python
%%time
for i in range(df.shape[0]):
    token_list = tokenizer(df['SentimentText'][i], stopwords_set)
    print(len(token_list))
    print()
    raw_data.append(token_list)
```

```python
print(len(raw_data))
print(df.shape[0])
```

# Build Vocabulary and Index2Word and Word2Index maps

```python
vocab_list = []
```

```python
%%time
vocab = set(vocab_list)
for i in range(len(raw_data)):
    vocab = vocab.union(set(raw_data[i]))
    
vocab_list = list(vocab)
```

```python
N = len(vocab_list)
print("Vocabulary Size: ", N)
```

```python
sorted_vocab_list = sorted(vocab_list)
```

```python
# Convert it to dictionary
vocab_dict_w2id = {}
vocab_dict_id2w = {}
for index,word in enumerate(sorted_vocab_list):
    vocab_dict_w2id[word]=index
    vocab_dict_id2w[index] = word
```

# Make raw count vectorizer

```python
raw_count = np.zeros((df.shape[0],N))
```

```python
%%time
for i in range(df.shape[0]):
    row_words_list = raw_data[i]
    for word in row_words_list:
        j = vocab_dict_w2id[word]
        raw_count[i][j] = raw_count[i][j]+1
        
```

```python
raw_count.shape
```

# Make Tf-Idf featurizer

```python
col_sum = raw_count.sum(axis=0)
col_sum.shape
```

```python
tf_idf = np.zeros((df.shape[0],N))
```

```python
%%time
for i in range(df.shape[0]):
    for j in range(N):
        tf_idf[i][j] = raw_count[i][j]*np.log(df.shape[0]/col_sum[j])
```
